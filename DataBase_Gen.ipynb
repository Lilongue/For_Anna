```python
start_date = dt.date(2019, 1, 19)
zakupki = pd.DataFrame({
    "ID": ["1"],
    "Дата": [str(start_date)],
    "Колличество":[1],
    "Стоимость": [100.01]})
spisanie = pd.DataFrame({
    "Заказ": ["1"],
    "ID": ["1"],
    "Дата": [str(start_date)],
    "Колличество": [1],
    "Стоимость": [100.1]})
spisanie
```

```python
start_date_o = start_date.toordinal()
stop_date = dt.date(2019, 12, 31)
stop_date_o = stop_date.toordinal()
print (start_date_o, stop_date_o)
```

```python
current_date_o = start_date_o
def get_zakaz(curr_date):
    ever_zakaz = int((curr_date - start_date_o) * (len(lst_zakaz))/(stop_date_o - start_date_o))
    rand_zakaz = random.randint(max(0,ever_zakaz-10), min(len(lst_zakaz) -1,ever_zakaz+10))
    return lst_zakaz[rand_zakaz]
def get_material():
    mater = materials["ID"][random.randint(0,len(materials["ID"]) - 1)]
    colvo = random.randint(1,5)
    return (mater, colvo)
def get_buy ():
    out_list = []
    for index, serie in materials.iterrows():
        if serie["Колличество"] < 10:
            out_list.append(serie["ID"])
    return out_list

mat_lst = []
zak_lst = []
date_lst = []
while current_date_o < stop_date_o:
    current_date_o += random.randint(1,2)
    td = dt.date.fromordinal(current_date_o)
    if td.weekday() >= 5:
        continue
    tmp_zakaz = get_zakaz(current_date_o)
    mat_colvo = random.randint(3,60)
    temp_set = set()
    for i in range(mat_colvo):
        mat = get_material()
        if mat[0] not in temp_set:
            temp_set.add(mat[0])
            mat_lst.append(mat)   
            zak_lst.append(tmp_zakaz)
            date_lst.append(str(td))
```

```python
for out in zip(zak_lst, mat_lst, date_lst):
    spisanie.loc[len(spisanie)] = [out[0], out[1][0], out[2], out[1][1], 1]
spisanie["Цена"] = 1
spisanie = spisanie[1:]
spisanie.head()
```

```python
for i, row in materials.iterrows():
    spisanie.loc[spisanie.ID == row["ID"], "Цена"] = row["Цена"]
spisanie["Стоимость"] = spisanie["Колличество"] * spisanie["Цена"]
spisanie.to_excel("spisanie.xlsx")
spisanie.head()
```
